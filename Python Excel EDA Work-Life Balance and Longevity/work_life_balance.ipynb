```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "/Users/laingangiang/Downloads/Work-Life Balance and Longevity.xlsx"
```

<h1>1. Average Age at Death by Occupation<h1>

```python
df1 = pd.read_excel(file_path, sheet_name="Sheet 1", header=2)
df1.columns = df1.columns.str.strip()
```

```python
print("The first few columns:")
df1.head()
```

```python
print("The last few columns:")
df1.tail()
```

```python
df1.describe()
```

```python
# Filter out the grand total row
df1 = df1[df1["Occupation type"] != "Grand Total"]

# Prepare data for plotting
occupations = df1["Occupation type"]
values = df1["Average of age at death"]

plt.subplots_adjust(bottom=0.3, left=0.08, right=0.98, top=0.9)

# Create the bar chart
plt.figure(figsize=(15, 7.5))
plt.xticks(rotation=45)

# Plot the bars with green color
plt.bar(occupations, values, color="#009E73")

# Add labels and title
plt.xlabel("Occupation type")
plt.ylabel("Average of age at death")
plt.title("Average Age at Death by Occupation", fontsize=12)

# Set y-axis limits
plt.ylim(60, 90)

# Adjust layout and display the chart
plt.tight_layout()
plt.show()
```

<h2>Average Rest, Sleep & Work Hours per Day by Occupation Type</h2>

```python
df2 = pd.read_excel(file_path, sheet_name="Sheet 2", header=2)
df2.columns = df2.columns.str.strip()
```

```python
print("The first few columns:")
df2.head()
```

```python
# Extract the relevant columns from df2
rest_hours = df2["Average of rest hours per day"]
sleep_hours = df2["Average of sleep hours per day"]
work_hours = df2["Average of work hours per day"]

# Create a 1x3 subplot layout for the three boxplots
fig, axs = plt.subplots(1, 3, figsize=(15, 7.5))

fig.subplots_adjust(bottom=0.3, left=0.08, right=0.98, top=0.9)

# First subplot: average rest hours
axs[0].boxplot(rest_hours, labels=["Average of rest hours"])
axs[0].set_title("Average of rest hours per day")
axs[0].set_ylabel("Average of rest hours")

# Second subplot: average sleep hours
axs[1].boxplot(sleep_hours, labels=["Average of sleep hours"])
axs[1].set_title("Average of sleep hours per day")
axs[1].set_ylabel("Average of sleep hours")

# Third subplot: average work hours
axs[2].boxplot(work_hours, labels=["Average of work hours"])
axs[2].set_title("Average of work hours per day", fontsize=12)
axs[2].set_ylabel("Average of work hours")

# Adjust spacing and display the figure
plt.tight_layout()
plt.show()
```

<h1>Average Age at Death by Occupation and Gender</h1>

```python
df3 = pd.read_excel(file_path, sheet_name="Sheet 3", header=2)
df3.columns = df3.columns.str.strip()
```

```python
print("The first few columns:")
df3.head()
```

```python
print("The last few columns:")
df2.tail()
```

```python
df3.describe()
```

```python
# Prepare the data for the grouped bar chart
occupation_type = df3["Occupation type"]
values1 = df3["Female"].sort_values(ascending=False)
values2 = df3["Male"]

# Define the x-axis positions and bar width
x = np.arange(len(occupation_type)) * 1.2
width = 0.5

# Create the figure and axis
fig, ax = plt.subplots(figsize=(15, 7.5))
fig.subplots_adjust(bottom=0.3, left=0.08, right=0.98, top=0.9)

# Plot the Female and Male bars side by side
bar1 = ax.bar(x - width/2, values1, width, label="Female", color="#56B4E9")
bar2 = ax.bar(x + width/2, values2, width, label="Male", color="#CC79A7")

# Add value labels to each bar
ax.bar_label(bar1, fmt="%.1f", padding=3)
ax.bar_label(bar2, fmt="%.1f", padding=3)

# Format the x-axis labels
ax.set_xticks(x)
ax.set_xticklabels(occupation_type, rotation=45, ha="right",)

# Add labels, title, and legend
ax.set_ylabel("Average Age at Death", fontsize=16)
ax.set_title("Average Age at Death by Occupation and Gender", fontsize=12)
ax.tick_params(axis="y", labelsize=12)
ax.tick_params(axis="x", labelsize=12)
ax.legend(fontsize=12)

# Set the y-axis limits
ax.set_ylim(50, 90)

# Display the chart
plt.tight_layout()
plt.show()
```

<h1>Average Work Hours per Day by Occupation and Gender</h1>

```python
df4 = pd.read_excel(file_path, sheet_name="Sheet 7", header=2)
df4.columns = df4.columns.str.strip()
```

```python
print("The first few columns:")
df4.head()
```

```python
filtered = df4[df4["Gender"] != "Grand Total"]

gender = filtered["Gender"]
bar1 = filtered["Average of rest hours per day"]
bar2 = filtered["Average of sleep hours per day"]
bar3 = filtered["Average of work hours per day"]

x = np.arange(len(gender))*2
width = 0.5

# 3. Create the plot figure
fig, ax = plt.subplots(figsize=(10, 5))

rects1 = ax.bar(x - width, bar1, width, label="Rest hours", color="#8CC5E3")
rects2 = ax.bar(x, bar2, width, label="Sleep hours", color="#298C8C")
rects3 = ax.bar(x + width, bar3, width, label="Work hours", color="#EA801C")

ax.set_xlabel("Gender")
ax.set_ylabel("Hours")
ax.set_title("Average Rest, Sleep & Work Hours per Day by Gender", fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(gender)
ax.legend()

# 6. Display the graph
plt.tight_layout()
plt.show()
```
